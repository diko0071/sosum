```python
from ntscraper import Nitter


scraper = Nitter(log_level=1, skip_instance_check=False)
```

```python
tweets = scraper.get_profile_info("korzhov_dm", mode='detail')
```

```python
tweets
```

```python
following_count = len(tweets['followers_list'])
followers_count = len(tweets['following_list'])
print(f"Following count: {following_count}, Followers count: {followers_count}")
```
